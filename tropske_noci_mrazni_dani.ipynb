# Racunanje linearne regresije za indeks broj tropskih noci

```python
import numpy as np
import matplotlib.pyplot as plt

#Ucitavanje fajla
ifile=np.loadtxt('obs_tasmin_vre_1950-2020_d_eobs_srb.txt')
```

### Definisanje funkcije linearna regresija

## NASLOV

```python
# Definisanje funkcije
def linearna_regresija(years, data):
    coefficients=np.polyfit(years, data, 1)
    k,n=coefficients[0],coefficients[1]
    lin_reg=k*years+n
    return lin_reg,k,n
```

### Nastavak programa: racunanje broja tropskih noci

```python
#Izvlacenje prve i cetvrte kolone
years=ifile[:,0]
tm=ifile[:,3]

ys=1950
ye=2020

#inicijalizacija liste 
no_tr=[] #broj tropskih noci
br=0
prag=20.

for i in range(ys, ye+1):
    for j in range(366):
        if years[np.where(years==i)[0][0]+j]==i and tm[np.where(years==i)[0][0]+j]>prag:
            br=br+1
    no_tr.append(br)
    br=0
    
no_tr=np.array(no_tr)
godine=np.array(range(ys,ye+1))

print('Broj tropskih noci po godinama')
print(godine)
print(no_tr)

```

### Crtanje linearne regresije

```python
#pozivanje funkcije
lin_reg,k,n=linearna_regresija(godine, no_tr)
lin_reg_5080,k1,n1 = linearna_regresija(godine[:30], no_tr[:30])
lin_reg_8120,k2,n2, = linearna_regresija(godine[31:], no_tr[31:])


plt.plot(godine, no_tr, 'b-', label='Broj tropskih noci')
plt.plot(godine, lin_reg, 'r.', label=f'Regresija 1950-2020: {k:.2f}*x+{n:.2f}')

plt.plot(godine[:30], lin_reg_5080, 'g.', label=f"Regresija 1950-1980: {k1:.2f}*x+{n1:.2f}")
plt.plot(godine[31:], lin_reg_8120, 'g.', label=f"Regresija 1981-2020: {k2:.2f}*x+{n2:.2f}")

plt.legend()
plt.title('Broj tropskih noci')
plt.xlabel('Godina')
plt.ylabel('tr')
```
